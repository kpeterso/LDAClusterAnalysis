```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
df=pd.read_csv('data.csv', sep=',')
df=df.drop('Respondent Number',1)
X=df.values
```

```python
df.tail()
```

```python
pd.get_dummies(df['Age'], prefix='Age').head()
```

```python
df = pd.concat([df, pd.get_dummies(df['Age'], prefix='Age')], axis=1)
df = df.drop(['Age'], axis=1)
df.head()
```

```python
range_n_clusters = [7, 8, 9, 10, 11, 12, 13]
for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,"The average silhouette_score is :", silhouette_avg)
    
```

```python
best_n=9
best_clusterer=KMeans(n_clusters=best_n, random_state=10)
cluster_labels=best_clusterer.fit_predict(X)
clusterCenters=best_clusterer.cluster_centers_
```

```python
clusterFrame=pd.DataFrame(cluster_labels)

centersFrame=pd.DataFrame(clusterCenters)
outData=pd.concat([df,clusterFrame])
```

```python
clusterFrame.head(n=5)
```

```python
clusterFrame.to_csv("ClusterLabels.csv", sep=',')
centersFrame.to_csv("ClusterCenters.csv", sep=',')
```
