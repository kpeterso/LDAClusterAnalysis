```python
from sklearn import linear_model
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn import datasets, svm
from sklearn.feature_selection import SelectPercentile, f_classif
```

```python
logistic = linear_model.LogisticRegression()
```

```python
df_all=pd.read_csv('Data/Midsize SUV Regression data.csv', sep=',')
df_all.head()
```

```python
X=df_all.values[:,9:]
y=df_all.values[:,8]
y = np.array(y).astype(int)
print(y[:10])
print(X[0,0])
df_all.head()
```

```python
#feature selection
plt.figure(1)
plt.clf()
X_indices = np.arange(X.shape[-1])
selector = SelectPercentile(f_classif, percentile=10)
selector.fit(X, y)
scores = -np.log10(selector.pvalues_)
scores /= scores.max()
plt.bar(X_indices - .45, scores, width=.2,label=r'Univariate score ($-Log(p_{value})$)', color='g')
plt.xlabel('Feature number')
plt.legend(loc='upper right')
plt.show()
```

```python
df_trunc=df_all.dropna()
df_trunc.head()
```

```python
r4=df_trunc['Demo_Census_4'].values
region4=np.unique(r4).tolist()
print(region4)
r9=df_trunc['Demo_Census_9'].values
region9=np.unique(r9).tolist()
```

```python
seriesList=df_all['MakeSeries'].unique()
print(seriesList)
#makeList=df_all['Make'].unique()
#print(makeList)
```

```python
frames=[]
keys=[]
X=df_all.values[:,9:]
y=df_all.values[:,8]
y = np.array(y).astype(int)
logistic.fit(X,y)
frames.append(pd.DataFrame(logistic.coef_))
keys.append('all')
for cluster in xrange(9):
    train_data = df_all.loc[df_all['Cluster']==cluster]
    X=train_data.values[:,9:]
    y=train_data.values[:,8]
    y = np.array(y).astype(int)
    logistic.fit(X,y)
    frames.append(pd.DataFrame(logistic.coef_))
    keys.append(cluster)
    
#for make in makeList:
#    train_data = df_all.loc[df_all['Make']==make]
#    X=train_data.values[:,7:]
#    y=train_data.values[:,6]
#    y = np.array(y).astype(int)
#    logistic.fit(X,y)
#    frames.append(pd.DataFrame(logistic.coef_))
#    keys.append(make)

for series in seriesList:
    train_data = df_all.loc[df_all['MakeSeries']==series]
    X=train_data.values[:,9:]
    y=train_data.values[:,8]
    y = np.array(y).astype(int)
    logistic.fit(X,y)
    frames.append(pd.DataFrame(logistic.coef_))
    keys.append(series)
    


for region in region4:
    train_data = df_trunc.loc[df_all['Demo_Census_4']==region]
    X=train_data.values[:,9:]
    y=train_data.values[:,8]
    y = np.array(y).astype(int)
    logistic.fit(X,y)
    frames.append(pd.DataFrame(logistic.coef_))
    keys.append(region)

for region in region9:
    train_data = df_trunc.loc[df_all['Demo_Census_9']==region]
    X=train_data.values[:,9:]
    y=train_data.values[:,8]
    y = np.array(y).astype(int)
    logistic.fit(X,y)
    frames.append(pd.DataFrame(logistic.coef_))
    keys.append(region)
    
result=pd.concat(frames,keys=keys).to_csv("results/Midsize SUV LogReg_coefs.csv")
```
