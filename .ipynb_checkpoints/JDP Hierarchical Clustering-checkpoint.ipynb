```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
from sklearn.metrics import silhouette_samples, silhouette_score
%pylab inline
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
df_all=pd.read_csv('Data/JDP Clustering data.csv', sep=',')
df_all.head()
```

```python
dummyVars=['Ownership Length','Miles / Day','YNV_GLB_3','YNV_GLB_4','YNV_GLB_5','YNV_GLB_6','YNV_GLB_7','YNV_GLB_8','YNV_GLB_9','DEMO_US_1A','DEMO_US_1B','DEMO_US_1C','DEMO_US_1D','DEMO_US_1E','DEMO_US_1F','DEMO_US_1G','DEMO_US_1H','DEMO_US_1I','DEMO_US_1J','DEMO_US_1K','DEMO_US_1L','Demo_AgeBucket','DEMO_INCM_1','Trim']
for d in dummyVars:
    df_all = pd.concat([df_all, pd.get_dummies(df_all[d], prefix=d)], axis=1)
    df_all = df_all.drop([d],axis=1)
```

```python
df_all.to_csv('transformed_JDP_clustering_data.csv')
df_all.head()
```

```python
segments=['Compact Car','Midsize Car','Compact SUV','Midsize SUV']
for segment in segments:
    train_data = df_all.loc[df_all['Segment']==segment]
    X=train_data.values[:,6:]
    #range_n_clusters = [5,6,7,8,9,10,11,12]
    #for n_clusters in range_n_clusters:
    n_clusters=9
    clusterer = AgglomerativeClustering(n_clusters=n_clusters)
    cluster_labels = clusterer.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For Segment ",segment,", n_clusters = ",n_clusters," the average silhouette_score is :", silhouette_avg)
    filename=segment+'_cluster_labels.csv'
    df_cluster_labels=pd.DataFrame(cluster_labels)
    df = pd.concat([pd.DataFrame(train_data['ID']).reset_index(drop=True), df_cluster_labels], axis=1).to_csv(filename)
```
