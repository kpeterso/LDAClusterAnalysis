```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('Data/SV 2016 correlation data.csv', sep=',')
```

```python
df.head()
```

```python
plt.matshow(df.corr())
df.corr().to_csv('results/overall correlation matrix.csv')
```

```python
sedanCorr=df.loc[(df['Match Segment'] == 'Sedan')].corr()
plt.matshow(df.loc[(df['Match Segment'] == 'Sedan')].corr())
sedanCorr.to_csv('results/sedan correlation matrix.csv')
```

```python
suvCorr=df.loc[(df['Match Segment'] == 'SUV')].corr()
plt.matshow(df.loc[(df['Match Segment'] == 'SUV')].corr())
suvCorr.to_csv('resluts/suv correlation matrix.csv')
```

```python
plt.matshow(df.loc[(df['Match Segment'] == 'Sedan')].corr()-df.loc[(df['Match Segment'] == 'SUV')].corr())
```

```python
delta=df.loc[(df['Match Segment'] == 'Sedan')].corr()-df.loc[(df['Match Segment'] == 'SUV')].corr()
```
