```python
from sklearn.cluster import AgglomerativeClustering
from scipy.stats import ttest_ind_from_stats
import numpy as np
import pandas as pd
from itertools import product
#%pylab inline
#import matplotlib.pyplot as plt
```

```python
#Load in data
df=pd.read_csv('Data/Edited UUNS.csv', sep=',')
df=df.drop(["XID"],axis=1)
#Do separate zip code analysis
df=df.drop(["Zip_Code"],axis=1)
num_category_cols=59
#df.dtypes
```

```python
print(df.columns.values.tolist()[59:62])
```

```python
#force data in first columns to be categorical
df_typed=df.iloc[:,0].astype('category')
#df_typed=pd.concat([df.iloc[:,:45].astype('category'), df.iloc[:,45:].astype('numeric')])
for i in range(1,num_category_cols+1):
    df_typed=pd.concat([df_typed,df.iloc[:,i].astype('category')],axis=1)

for i in range(num_category_cols+1,532):
    df_typed=pd.concat([df_typed,df.iloc[:,i].astype('float')],axis=1)

#df_typed.describe()
```

```python
df_typed.shape
```

```python
#Create list of lists of category values for each category
d={}
nameList=['all']
for i in range(num_category_cols):
    name=df_typed.iloc[:,i].name
    v=df_typed.iloc[:,i].cat.categories.tolist()
    d[name]=v
    
#create mean, standard deviation and sample size dataframes. Initialize with full dataset.
df_mean=df_typed.mean(numeric_only=True).to_frame()
df_std=df_typed.mean(numeric_only=True).to_frame()
df_count=df_typed.count(numeric_only=True).to_frame()
df_names=

#rename columns of array        
df_mean.columns = nameList
df_std.columns = nameList
df_count.columns = nameList

#write array to output
df_mean.to_csv('count.csv')

#loop through the categories, isolate each category value
for key in d.iterkeys():
    valueList=d[key]
    for j in valueList:
        #find mean, stDev, and sample size for each column
        df_mean=pd.concat([df_mean,df_typed[df_typed[key]==j].mean(numeric_only=True).to_frame()],axis=1)
        df_std=pd.concat([df_std,df_typed[df_typed[key]==j].std(numeric_only=True).to_frame()],axis=1)
        df_count=pd.concat([df_count,df_typed[df_typed[key]==j].count(numeric_only=True).to_frame()],axis=1)
        nameList.append('_'.join((key,str(j))))

#rename columns of array        
df_mean.columns = nameList
df_std.columns = nameList
df_count.columns = nameList
#df_count=df_count.fillna(0)
```

```python
#create numpy arrays of values
meanArray=df_mean.values
countArray=df_count.values
stdArray=df_std.values

df_mean.ix[df_count.my_channel > 20000, 'my_channel'] = 0
```

```python
#get column titles in a list
titles=['drop this']
column_header_list=df_mean.columns.values.tolist()

mean_array=df_mean.values
count_array=df_count.values
std_array=df_std.values

#initialize the significance and p-value dataframes
t,p=ttest_ind_from_stats(mean_array[:,0],std_array[:,0],count_array[:,0],mean_array[:,1],std_array[:,1],count_array[:,1],equal_var=False)
df_t=pd.DataFrame(t)
df_p=pd.DataFrame(p)

#loop through ever combination of columns, test significance
for i in range(df_mean.shape[1]-1):
    for j in range(i+1,df_mean.shape[1]):
        #use wards significance test to measure significance, returns t-score and p-value
        t,p=ttest_ind_from_stats(mean_array[:,i],std_array[:,i],count_array[:,i],mean_array[:,j],std_array[:,j],count_array[:,j],equal_var=False)
        df_t=pd.concat([df_t,pd.DataFrame(t)],axis=1)
        df_p=pd.concat([df_p,pd.DataFrame(p)],axis=1)
        #get title name
        titles.append(';'.join([column_header_list[i],column_header_list[j]]))
        
df_p.columns=titles
df_t.columns=titles
df_t.index=df_mean.index
df_p.index=df_mean.index
df_t = df_t.ix[1:]
df_p=df_p.ix[1:]
df_t=df_t.drop(['drop this'],axis=1)
df_p=df_p.drop(['drop this'],axis=1)
```
