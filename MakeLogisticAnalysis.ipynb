```python
from sklearn import linear_model
import pandas as pd
import numpy as np
```

```python
logistic = linear_model.LogisticRegression()
```

```python
df_MS=pd.read_csv('Data/Midsize SUV Regression data.csv', sep=',')
df_MS = df_MS.drop(['Demo_Census_4'],axis=1)
df_MS = df_MS.drop(['Demo_Census_9'],axis=1)
df_all.head()
```

```python
df_MC=pd.read_csv('Midsize Car Regression data.csv', sep=',')
df_MC = df_MC.drop(['Demo_Census_4'],axis=1)
df_MC = df_MC.drop(['Demo_Census_9'],axis=1)
df_all.head()
```

```python
df_CS=pd.read_csv('Compact SUV Regression data.csv', sep=',')
df_CS = df_CS.drop(['4Region'],axis=1)
df_CS = df_CS.drop(['9Region'],axis=1)
df_all.head()
```

```python
df_CC=pd.read_csv('Compact Car Regression data.csv', sep=',')
df_all.head()
```

```python
l=[df_MC,df_MS,df_CS,df_CC]
df_all = pd.concat(l, axis=0)
df_all=df_all.drop("APEAL_Index",1)
df_all.tail()
```

```python
makeList=df_all['Make'].unique()
print(makeList)
```

```python
frames=[]
keys=[]
X=df_all.values[:,7:]
y=df_all.values[:,6]
y = np.array(y).astype(int)
logistic.fit(X,y)
frames.append(pd.DataFrame(logistic.coef_))
keys.append('all')
    
for make in makeList:
    train_data = df_all.loc[df_all['Make']==make]
    X=train_data.values[:,7:]
    y=train_data.values[:,6]
    y = np.array(y).astype(int)
    logistic.fit(X,y)
    frames.append(pd.DataFrame(logistic.coef_))
    keys.append(make)
    
result=pd.concat(frames,keys=keys).to_csv("results/Make LogReg_coefs.csv")
```
